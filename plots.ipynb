This notebook is used to plot histograms and scatter plots of features in the honey production and air quality datasets.<br>
First, imports and read in the data

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

honey_data = pd.read_csv('data/honeyproduction.csv')
feature_vectors = pd.read_csv('data/completeFeatureVectors.csv')
```

```python
sns.histplot(data=honey_data, x="yieldpercol").set_title("Distribution of Honey Yields per Colony in the US 1998-2012")
```

```python
sns.histplot(data=honey_data, x="year", bins=15).set_title("Number of States with Honey Production Data Per Year 1998-2012")
```

```python
fig, axs = plt.subplots(ncols=4, nrows=2, figsize=(20,8))
sns.scatterplot(data=feature_vectors[feature_vectors["pm25_frm"] < 20], x="pm25_frm", y="yield_per_col", ax=axs[0,0])
sns.scatterplot(data=feature_vectors, x="o3", y="yield_per_col", ax=axs[1,0])
sns.scatterplot(data=feature_vectors, x="co", y="yield_per_col", ax=axs[0,1])
sns.scatterplot(data=feature_vectors, x="so2", y="yield_per_col", ax=axs[1,1])
sns.scatterplot(data=feature_vectors, x="no2", y="yield_per_col", ax=axs[0,2])
sns.scatterplot(data=feature_vectors[feature_vectors["pressure"] < 1200], x="pressure", y="yield_per_col", ax=axs[1,2])
sns.scatterplot(data=feature_vectors, x="temperature", y="yield_per_col", ax=axs[0,3])
sns.scatterplot(data=feature_vectors, x="wind", y="yield_per_col", ax=axs[1,3])
```
