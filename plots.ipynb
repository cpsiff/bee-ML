This notebook is used to plot histograms and scatter plots of features in the honey production and air quality datasets.<br>
First, imports and read in the data

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

honey_data = pd.read_csv('honeyproduction.csv')
```

```python
sns.histplot(data=honey_data, x="yieldpercol").set_title("Distribution of Honey Yields per Colony in the US 1998-2012")
```

```python
sns.histplot(data=honey_data, x="year", bins=15).set_title("Number of States with Honey Production Data Per Year 1998-2012")
```
